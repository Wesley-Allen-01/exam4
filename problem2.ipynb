```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1 = [5, 0, 1]
x2 = [0, 5, -1]
x3 = [5, 5, 1]
plt.scatter([x1[0], x2[0], x3[0]], [x1[1], x2[1], x3[1]])

plt.axvline(x=2.5, color='r', linestyle='--')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# plot the function x = 2.5 on the same graph
x = np.linspace(-10, 10, 100)
y = [2.5] * 100

plt.plot(x, y)
```
